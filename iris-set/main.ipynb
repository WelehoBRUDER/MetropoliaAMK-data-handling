# Iris set sample
Data loaded from "iris.csv"

```python
import pandas as pd

df = pd.read_csv('iris.csv')

df.info()
```

## Frame manipulation
Splitting off a part of the frame

```python
df_part = df.loc[[10, 20, 30], :]

# Refers to pandas index
df_part.loc[10]

# Refers to python index
df_part.iloc[0]

# Both lines do the same thing but with different methods
```

## Further manipulation
Example: multiply all values in column by 10

```python
df["sepal_length"] = df["sepal_length"] * 10

df
```

```python
# Revert the change and calc mean
df["sepal_length"] = df["sepal_length"] / 10

df.describe()
```
